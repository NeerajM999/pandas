```python
import pandas as pd
```

```python
chicago = pd.read_csv("data/chicago.csv")
chicago.head(5)
```

```python
chicago.info()
```

```python
# which columns we can optimize for storage and memory usage. Department has only 35 values, it can be optimized.
chicago.nunique()
```

```python
chicago['Department'] = chicago['Department'].astype('category')
chicago.info()
```

```python
# convert to lower case
chicago['Name'].str.lower().str.upper().head(5)
```

```python
# convert to title case
chicago['Name'].str.title().head(4)
```

```python
# get length of each record
chicago['Name'].str.len().head(5)
```

```python
# replace method, lets replace $ in salary and convert it to float
chicago['Employee Annual Salary'] = chicago['Employee Annual Salary'].str.replace('$', '').astype(float).head(5)
```

```python
chicago.head(5)
```

```python
# we can also replace columns
chicago.columns = chicago.columns.str.replace(' ', '_')
chicago.head(5)
```

```python
# filtering
# we must remove NA or NaN values before this
chicago =  pd.read_csv("data/chicago.csv").dropna(how='all')
chicago['Department'] = chicago['Department'].astype('category')
mask = chicago['Position Title'].str.lower().str.contains('water')
chicago[mask].head(5)
```

```python
# starts with
mask = chicago['Position Title'].str.lower().str.startswith("water")
chicago[mask].head(5)
```

```python
# ends with
mask = chicago['Position Title'].str.lower().str.endswith('ist')
chicago[mask].head(5)
```

```python
# remove white space from strings
"   Hello ".lstrip()
```

```python
"   Hello  ".rstrip()
```

```python
"  Hello  ".strip()
```

```python
chicago['Name'].str.strip().head(5)
```

```python
chicago.index = chicago['Name']
```

```python
chicago = chicago.drop(columns='Name')
```

```python
chicago.index
```

```python
chicago.index = chicago.index.str.title().str.strip()
```

```python
chicago.index
```

```python
chicago.head(5)
```

```python
# split
chicago.reset_index(inplace=True)
```

```python
chicago.head(5)
```

```python
chicago['Name'].str.split(",").str.get(0).value_counts().head(10)
```

```python
# get most common first word in position title
chicago['Position Title'].str.split(" ").str.get(0).value_counts().head(5)
```

```python
# get most common first names
chicago['Name'].str.split(",").str.get(1).str.strip().str.split(" ").str.get(0).value_counts().head(5)
```

```python
# get least common first names
chicago['Name'].str.split(",").str.get(1).str.strip().str.split(" ")\
    .str.get(0).value_counts().sort_values(ascending=True).head(5)
```

```python
# expand method. It splits and generates dataframe instead of list of arrays
chicago['Name'].str.split(",", expand=True).head(5)
```

```python
# now we can add 2 columns one for each part of the name
chicago[["fname", "lname"]] = chicago['Name'].str.split(",", expand=True)
chicago.head(5)
```

```python
chicago['Position Title'].str.split(" ", expand=True)
```

```python
# we can limit the no. of splits needed in above case
chicago['Position Title'].str.split(" ", expand=True, n=1)
```

```python
# aggregations or grouping

foods = pd.read_csv("data/foods.csv")
foods.head(5)
```

```python
# average spending per gender
foods.pivot_table(values='Spend', index='Gender', aggfunc='mean')
```

```python
# total spending per gender
foods.pivot_table(values='Spend', index='Gender', aggfunc='sum')
```

```python
# total spend per gender and item
foods.pivot_table(values='Spend', index=['Gender', 'Item'], aggfunc='sum')
```

```python
# total spend per gender and item and city
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns = "City", aggfunc='sum')
```

```python
# total spend per gender and item and city
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns = ["City","Frequency"], aggfunc='sum')
```
