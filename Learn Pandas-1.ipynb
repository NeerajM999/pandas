```python
# read a xml file into pandas dataframe
import xml.etree.ElementTree as ET
import io

def iter_docs(books):
    book_attr = books.attrib
    for doc in books.iter('book'):
        doc_dict = book_attr.copy()
        doc_dict.update(doc.attrib)
        doc_dict['data'] = doc.text
        yield doc_dict
        
etree = ET.parse("data/books.xml")
doc_df = pd.DataFrame(list(iter_docs(etree.getroot())))

doc_df.head()
    
```

# Merging, Joining and concatenating

```python
week1 = pd.read_csv("data/Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("data/Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("data/Restaurant - Customers.csv")
foods = pd.read_csv("data/Restaurant - Foods.csv")
```

```python
# concatenate or combine multiple dataframes
sales = pd.concat([week1, week2], ignore_index=True)
len(sales)
```

```python
# what if you want to retain the key about where the record came from in final dataframe
sales = pd.concat([week1, week2], keys=["Week 1", "Week 2"])
sales
```

```python
sales.loc["Week 1"]
```

```python
# get a specific element
sales.loc[("Week 1", 240)]
```

```python
sales.loc[("Week 1", 240), "Customer ID"]
```

```python
# another approach for concat is append
week1.append(week2, ignore_index=True)
```

```python
# inner join
week1.merge(week2, how="inner", on="Customer ID")
```

```python
# get customers who ordered same food in both weeks
week1.merge(week2, how="inner", on=["Customer ID", "Food ID"])
```

```python
# validate
week1[week1["Customer ID"] == 578]
```

```python
week2[week2["Customer ID"] == 578]
```

```python
# outer join
merged = week1.merge(week2, how="outer", on="Customer ID", indicator=True)
merged
```

```python
merged['_merge'].value_counts()
```

```python
mask = merged["_merge"].isin(["left_only", "right_only"])
merged[mask]
```

```python
# left join
week1.merge(foods, how="left", on="Food ID", sort = True)
```

```python
# when joining cols are different
week2.merge(customers, how="left", left_on="Customer ID", right_on="ID").drop("ID", axis="columns")
```

```python
# joining DFs on index and column
customers = pd.read_csv("data/Restaurant - Customers.csv" , index_col="ID")
customers.head(3)
```

```python
sales = week1.merge(customers, how="left", left_on="Customer ID", right_index=True)
sales.head(5)
```

```python
sales.merge(foods, how = "left", left_on="Food ID", right_index=True)
```

```python
# join dataframes vertically ex. on indexes
satisfaction = pd.read_csv("data/Restaurant - Week 1 Satisfaction.csv")
week1.merge(satisfaction, how="left", left_index = True, right_index=True)
```

```python
# get hours
times = pd.date_range(start='2019-02-10', end='2019-02-20', freq = '2H')
times
```
