```python
## convert a value in cells while loading
def convert_no_rain(cell):
    if cell == 0.0:
        return 'No Rain'
    else:
        return cell
    
def convert_wind(cell):
    if cell == 0.0:
        return 'No Wind'
    else:
        return cell

    
df = pd.read_excel('/Users/Neeraj_Malhotra/Downloads/weather-trends.xls', 
                   sheet_name='weather-trends', 
                  converters={
                      'yearly_ra': convert_no_rain,
                      'wind_speed': convert_wind})

df[df.yearly_ra == 'No Rain']
```

```python
# combining two dataframes sidewise
temp_df = pd.DataFrame({
    "city": ["mumbai", "delhi", "bangalore"],
    "temperature": [44,55,66]
})

temp_df

windspeed_df = pd.DataFrame({
    "city": ["delhi", "mumbai", "bangalore", 'pune'],
    "wind": [12,4,5, 55]
})
windspeed_df
```

```python
# this will give wrong results as index are out of order in both DFs... 
df = pd.concat([temp_df, windspeed_df], axis=1)
df
```

```python
# solution 1, set index to common column
dflist = [temp_df, windspeed_df]
dfs = [df.set_index('city') for df in dflist]
print(pd.concat(dfs, axis=1))
```

```python
# solution#2 using merge
# default is inner join which will drop any non matching rows
df = pd.merge(temp_df, windspeed_df, on="city")
df
```

```python
# so we need to use outer join
df = pd.merge(temp_df, windspeed_df, on="city", how="outer")
df
```
